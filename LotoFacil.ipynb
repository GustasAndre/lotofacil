```python
#Funções importantes
# Contagem de números primos por jogo
def contar_primos(numeros):
    primos = 0
    for num in numeros:
        if num > 1:
            for i in range(2, num):
                if (num % i) == 0:
                    break
            else:
                primos += 1
    return primos

#Contar os numeros sequenciais
def maior_sequencia(numeros, limite):
    max_sequencia = 0
    sequencia_atual = 0
    for i in range(len(numeros) - 1):
        if numeros[i + 1] - numeros[i] == 1:
            sequencia_atual += 1
            if sequencia_atual >= limite:
                sequencia_atual = limite - 1  # Limitar a sequência ao valor máximo
        else:
            max_sequencia = max(max_sequencia, sequencia_atual)
            sequencia_atual = 0
    return max(max_sequencia, sequencia_atual) + 1  # Adicionando 1 porque estamos considerando a sequência atual como não sequencial no loop
```

```python
# Criar características dos números sorteados
df_resultados['Soma_Numeros'] = df_resultados.loc[:, 'bola_1':'bola_15'].sum(axis=1)  # Soma dos números sorteados
df_resultados['Media_Numeros'] = df_resultados.loc[:, 'bola_1':'bola_15'].mean(axis=1)  # Média dos números sorteados
df_resultados['Pares'] = df_resultados.loc[:, 'bola_1':'bola_15'].apply(lambda row: sum(1 for num in row if num % 2 == 0), axis=1)  # Quantidade de números pares
df_resultados['Impares'] = df_resultados.loc[:, 'bola_1':'bola_15'].apply(lambda row: sum(1 for num in row if num % 2 != 0), axis=1)  # Quantidade de números ímpares

# Exemplo de criação de uma nova característica com base na soma dos dois primeiros números sorteados
df_resultados['Soma_Primeiros_Numeros'] = df_resultados.apply(lambda row: row['bola_1'] + row['bola_2'] if len(row) >= 2 else None, axis=1)

df_resultados['Numeros_Primos'] = df_resultados.loc[:, 'bola_1':'bola_15'].apply(contar_primos, axis=1)

# Maior sequência de números por jogo, limitando a 5 números sequenciais
df_resultados['Maior_Sequencia'] = df_resultados.loc[:, 'bola_1':'bola_15'].apply(lambda row: maior_sequencia(sorted(row), 5), axis=1)
```

```python
def calcular_metricas(df, ultimos_concursos=10):
    # Números que mais saíram
    numeros_mais_sairam = df.loc[:, 'bola_1':'bola_15'].stack().value_counts().sort_values(ascending=False)

    # Números que menos saíram
    numeros_menos_sairam = df.loc[:, 'bola_1':'bola_15'].stack().value_counts().sort_values()

    # Números que estão saindo nos últimos concursos
    ultimos_concursos_df = df.tail(ultimos_concursos)
    numeros_saindo_ultimos_concursos = ultimos_concursos_df.loc[:, 'bola_1':'bola_15'].stack().value_counts()

    # Números que não estão saindo nos últimos concursos
    numeros_nao_saindo_ultimos_concursos = set(range(1, 26)) - set(numeros_saindo_ultimos_concursos.index)

    return numeros_mais_sairam, numeros_menos_sairam, numeros_saindo_ultimos_concursos, numeros_nao_saindo_ultimos_concursos

# Chamar a função para calcular as métricas
numeros_mais_sairam, numeros_menos_sairam, numeros_saindo_ultimos, numeros_nao_saindo_ultimos = calcular_metricas(df_resultados)
```
